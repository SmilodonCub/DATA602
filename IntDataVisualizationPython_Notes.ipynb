### Customizing Axes

Controlling axis extents
* `axis( [ xmin, xmax, ymin, ymax ] )`
* or control individual axis extants:
    + `xlim( [ xmin, xmax ] )`
    + `ylim( [ ymin, ymax ] )`
* can use either tuples or lists:
    + `xlim( ( -2,3 ) )` works
    + `xlim( [ -2,3 ] )` works all the same

```python
usgdp_url = 'https://raw.githubusercontent.com/SmilodonCub/DATA602/master/gdp_year.csv'
usgdp_df = pd.read_csv( usgdp_url )
usgdp_df.head()
```

```python
#using xlim & ylim to subset the plot
plt.plot( usgdp_df[ 'date' ], usgdp_df[ 'level-current'] )
plt.xlabel( 'Year' )
plt.ylabel( 'Billions USD' )
plt.title( 'US Gross Domestic Product' )
plt.xlim( ( 1947, 1957 ) )
plt.ylim( ( 0, 1000 ) )
plt.show()
```

```python
#a single call to .axis()
plt.plot( usgdp_df[ 'date' ], usgdp_df[ 'level-current'] )
plt.xlabel( 'Year' )
plt.ylabel( 'Billions USD' )
plt.title( 'US Gross Domestic Product' )
plt.axis( ( 1980, 2010, 0, 20000))
plt.show()
```

other `axis()` options:
* `axis('off')` turns off axis lines, labels
* `axis('equal')` equal scaling on x and y axis
* `axis('square')` forces square plot
* `axis('tight')` sets xlim, ylim to show all data

```python
an = np.linspace( 0,2*np.pi, 100 )
fig, axs = plt.subplots( 3,2 )

axs[0, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 0].set_title('not equal, looks like ellipse', fontsize=10)

axs[0, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 1].axis('equal')
axs[0, 1].set_title('equal, looks like circle', fontsize=10)

axs[1, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 0].axis('square')
axs[1, 0].set_title('still a circle, limits fit to data', fontsize=10)

axs[1, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 1].axis('tight')
axs[1, 1].set_title('not a circle, limits fit to data', fontsize=10)

axs[2, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[2, 0].axis('off')
axs[2, 0].set_title('axis off', fontsize=10)

axs[2, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[2, 1].set_aspect('equal', 'box')
axs[2, 1].axis('off')
axs[2, 1].set_title('axis off, circle again', fontsize=10)

fig.tight_layout()

plt.suptitle( 'Playing with axis()', y = 1.05)

plt.show()
```

## Working with 2D Arrays

Refresher: NumPy Arrays
* Homogeneous in data type
* Supports vectorized calculations
* Indexing with brackets:
    + `A[ index ] for a 1D array
    + `A[ index0, index1 ] for a 2D array

Reminder: slicing arrays
* slicing 1D arrays: `A[ index ]`, 2D arrays `A[ slice0, slice1 ]`
* slicing: `slice = start:stop:stride`
* indexes from `start` to `stop-1` in steps of `stride`
* missing start: implies the start of the array
* missing stop: implies the end of the array 
* missing stride: implies step of 1
* negative indices/slices: count from end of array

```python
u = np.linspace( -2,2,30 )
v = np.linspace( -1,1,50 )
X,Y = np.meshgrid( u,v )

Z= X**2/25 + Y**2/4

print( Z )
plt.set_cmap( 'gray' )
plt.pcolor( Z )
plt.show()
```

```python
# Import numpy and matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1,1,21,)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid( u,v )

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor( Z )
plt.show()
```

```python
A = np.array([[1, 0, -1], [2, 0, 1], [1, 1, 1]])
plt.pcolor(A, cmap='Blues')
plt.colorbar()
plt.show()
```

### Visualizing Bivariate Functions

```python
u = np.linspace( -2,2,65 )
v = np.linspace( -1,1,33 )
X,Y = np.meshgrid( u,v )

Z = np.cos(2*np.sqrt(X**2 + Y**2)) 


colormaps = [ 'inferno', 'magma', 'plasma', 'viridis' ]
for colormap in list( range( 4 ) ):
    #plt.plot( 4colormap+1 )
    plt.set_cmap( colormaps[ colormap ] )
    plt.pcolor( Z )
    plt.show()
```

```python
#Pseudocolor plot
plt.pcolor( X,Y,Z, cmap= 'autumn' ) #axes are in terms of variable values instead of index
plt.colorbar()
plt.axis('tight')
plt.show()
```

```python
Z = np.cos(2*np.sqrt(X**2 + Y**2)) 
plt.contour( X,Y,Z, 30 )
plt.show()
```

```python
plt.contourf( X,Y,Z, 20, cmap = 'prism' )
plt.colorbar()
plt.show()
```

```python
# Generate a default contour map of the array Z
plt.subplot(2,2,1)
plt.contour( X, Y, Z )

# Generate a contour map with 20 contours
plt.subplot(2,2,2)
plt.contour( X, Y, Z, 20 )

# Generate a default filled contour map of the array Z
plt.subplot(2,2,3)
plt.contourf( X, Y, Z, 20 )

# Generate a default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf( X, Y, Z, 20 )

# Improve the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Create a filled contour plot with a color map of 'viridis'
plt.subplot(2,2,1)
plt.contourf(X,Y,Z,20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

# Create a filled contour plot with a color map of 'gray'
plt.subplot(2,2,2)
plt.contourf(X,Y,Z,20, cmap='gray')
plt.colorbar()
plt.title('Gray')

# Create a filled contour plot with a color map of 'autumn'
plt.subplot(2,2,3)
plt.contourf(X,Y,Z,20, cmap='autumn')
plt.colorbar()
plt.title('Autumn')

# Create a filled contour plot with a color map of 'winter'
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20, cmap='winter')
plt.colorbar()
plt.title('Winter')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```
