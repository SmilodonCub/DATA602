### Customizing Axes

Controlling axis extents
* `axis( [ xmin, xmax, ymin, ymax ] )`
* or control individual axis extants:
    + `xlim( [ xmin, xmax ] )`
    + `ylim( [ ymin, ymax ] )`
* can use either tuples or lists:
    + `xlim( ( -2,3 ) )` works
    + `xlim( [ -2,3 ] )` works all the same

```python
usgdp_url = 'https://raw.githubusercontent.com/SmilodonCub/DATA602/master/gdp_year.csv'
usgdp_df = pd.read_csv( usgdp_url )
usgdp_df.head()
```

```python
#using xlim & ylim to subset the plot
plt.plot( usgdp_df[ 'date' ], usgdp_df[ 'level-current'] )
plt.xlabel( 'Year' )
plt.ylabel( 'Billions USD' )
plt.title( 'US Gross Domestic Product' )
plt.xlim( ( 1947, 1957 ) )
plt.ylim( ( 0, 1000 ) )
plt.show()
```

```python
#a single call to .axis()
plt.plot( usgdp_df[ 'date' ], usgdp_df[ 'level-current'] )
plt.xlabel( 'Year' )
plt.ylabel( 'Billions USD' )
plt.title( 'US Gross Domestic Product' )
plt.axis( ( 1980, 2010, 0, 20000))
plt.show()
```

other `axis()` options:
* `axis('off')` turns off axis lines, labels
* `axis('equal')` equal scaling on x and y axis
* `axis('square')` forces square plot
* `axis('tight')` sets xlim, ylim to show all data

```python
an = np.linspace( 0,2*np.pi, 100 )
fig, axs = plt.subplots( 3,2 )

axs[0, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 0].set_title('not equal, looks like ellipse', fontsize=10)

axs[0, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 1].axis('equal')
axs[0, 1].set_title('equal, looks like circle', fontsize=10)

axs[1, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 0].axis('square')
axs[1, 0].set_title('still a circle, limits fit to data', fontsize=10)

axs[1, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 1].axis('tight')
axs[1, 1].set_title('not a circle, limits fit to data', fontsize=10)

axs[2, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[2, 0].axis('off')
axs[2, 0].set_title('axis off', fontsize=10)

axs[2, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[2, 1].set_aspect('equal', 'box')
axs[2, 1].axis('off')
axs[2, 1].set_title('axis off, circle again', fontsize=10)

fig.tight_layout()

plt.suptitle( 'Playing with axis()', y = 1.05)

plt.show()
```

### Legends, Annotations and Styles

```python
from sklearn.datasets import load_iris
```

```python
iris_dat = load_iris() 
iris_df = pd.DataFrame( iris_dat.data, columns = iris_dat.feature_names )
iris_df.head()
```

```python
print( iris_dat.target_names )
print( set( iris_dat.target ) )
iris_df['target'] = iris_dat.target
iris_df.head()
```

```python
plt.subplot( 1,2,1 )
plt.scatter( iris_df[ iris_df['target']==0 ]['sepal length (cm)'],
            iris_df[ iris_df['target']==0 ]['sepal width (cm)'],
            marker='o', color = 'red', label='setosa' )
plt.scatter( iris_df[ iris_df['target']==1 ]['sepal length (cm)'],
            iris_df[ iris_df['target']==1 ]['sepal width (cm)'],
            marker='o', color = 'green', label='versicolor' )
plt.scatter( iris_df[ iris_df['target']==2 ]['sepal length (cm)'],
            iris_df[ iris_df['target']==2 ]['sepal width (cm)'],
            marker='o', color = 'blue', label='virginica' )
#plt.legend( loc = 'upper right' )
plt.title( 'Sepal dim' )
plt.xlabel( 'sepal length (cm)' )
plt.ylabel( 'sepal width (cm)' )
plt.xlim( ( 2,10 ) )
plt.ylim( ( 1,6 ) )

plt.subplot( 1,2,2 )
plt.scatter( iris_df[ iris_df['target']==0 ]['petal length (cm)'],
            iris_df[ iris_df['target']==0 ]['petal width (cm)'],
            marker='o', color = 'red', label='setosa' )
plt.scatter( iris_df[ iris_df['target']==1 ]['petal length (cm)'],
            iris_df[ iris_df['target']==1 ]['petal width (cm)'],
            marker='o', color = 'green', label='versicolor' )
plt.scatter( iris_df[ iris_df['target']==2 ]['petal length (cm)'],
            iris_df[ iris_df['target']==2 ]['petal width (cm)'],
            marker='o', color = 'blue', label='virginica' )
plt.legend( loc = 'upper right' )
plt.suptitle( 'Iris data' )
plt.title( 'Petal dim')
plt.xlabel( 'petal length (cm)' )
plt.ylabel( 'petal width (cm)' )
plt.xlim( ( 0,10 ) )
plt.ylim( ( 0,4 ) )

plt.show()
```

Plot Annotations
* test labels and arrows using `annotate()` method
* flexible specification of coordinates
* keyword `arrowprops`:`dict` of arrow properties
    + width
    + color
    + etc.

```python
plt.scatter( iris_df[ iris_df['target']==0 ]['sepal length (cm)'],
            iris_df[ iris_df['target']==0 ]['sepal width (cm)'],
            marker='o', color = 'red', label='setosa' )
plt.scatter( iris_df[ iris_df['target']==1 ]['sepal length (cm)'],
            iris_df[ iris_df['target']==1 ]['sepal width (cm)'],
            marker='o', color = 'green', label='versicolor' )
plt.scatter( iris_df[ iris_df['target']==2 ]['sepal length (cm)'],
            iris_df[ iris_df['target']==2 ]['sepal width (cm)'],
            marker='o', color = 'blue', label='virginica' )
#plt.legend( loc = 'upper right' )
plt.title( 'Sepal dim' )
plt.xlabel( 'sepal length (cm)' )
plt.ylabel( 'sepal width (cm)' )
plt.xlim( ( 4,8.5 ) )
plt.ylim( ( 1.5,5 ) )
plt.annotate( 'setosa', xy = ( 5.0, 3.5 ), size = 20, color = 'red' )
plt.annotate( 'versicolor', xy = ( 7.25, 3.5 ), size = 20, color='blue' )
plt.annotate( 'virginica', xy = ( 5.0, 2.0 ), size = 20, color = 'green' )
plt.show()
```
